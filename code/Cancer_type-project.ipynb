```python
import pydicom
import matplotlib.pyplot as plt
import os

# Ruta a la carpeta de la serie
folder_path = "C:\Users\mario\OneDrive\MAIA\semestre 4\PDS\Microproyecto\TCGA-CESC\TCGA-VS-A8EB\05-05-2004-NA-RESSONANCIA MAGNETICA DE ABDOME INFERIOR-20151\301.000000-AXI T2 ssSH 5MM-79572"

# Leer todos los archivos DICOM y ordenarlos por posición de corte
dicoms = [pydicom.dcmread(os.path.join(folder_path, f))
          for f in os.listdir(folder_path) if f.endswith(".dcm")]
dicoms.sort(key=lambda x: float(x.ImagePositionPatient[2]))  # Orden por eje Z

# Mostrar el corte central
mid_slice = dicoms[len(dicoms)//2]
plt.imshow(mid_slice.pixel_array, cmap='gray')
plt.title(mid_slice.SeriesDescription)
plt.axis('off')
plt.show()
```

```python
# Conteo absoluto y porcentual
counts = df_filtrado["Disease Type"].value_counts()
percentages = (counts / counts.sum() * 100).round(2)

# Crear gráfico de barras
ax = counts.plot(kind="bar", color="skyblue", figsize=(8,5))
plt.title("Distribución de Cancer Type Detailed en los 54 pacientes")
plt.xlabel("Subtipo de cáncer")
plt.ylabel("Número de pacientes")

# Agregar anotaciones con porcentajes arriba de cada barra
for i, (count, pct) in enumerate(zip(counts, percentages)):
    ax.text(i, count + 0.5, f"{pct}%", ha="center", va="bottom", fontsize=9, fontweight="bold")

plt.show()
```
